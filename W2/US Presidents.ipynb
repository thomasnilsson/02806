````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### Loading the ```.csv``` file with pandas

````python
data = pd.read_csv('data/presidents.csv').drop('number', 1)
months = data.months
mean = np.mean(months)
````

### Scatter plot for the president data (figure 2-1)

````python
plt.rcParams["figure.figsize"] = [6,4]
plt.plot(months, "o", label="Tenure in months")
plt.axhline(y=mean, linestyle='-', label = "Mean tenure = " + str(int(mean)) + " months")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
````

### Histogram of the president data (figure 2-4)

````python
plt.rcParams["figure.figsize"] = [20,4]
plt.hist(months, bins = max(months), edgecolor='black', linewidth=0.5)
plt.title("US Presidents Tenure")
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Presidents")
plt.show()
````

### CDF of president Data

````python
x = np.sort(months)
y = np.arange(1, len(months) + 1/len(months))/len(months)
plt.plot(x, y, '.', linestyle='none')
plt.title("US Presidents Tenure")
plt.xlabel("Tenure (Months)")
plt.ylabel("% of Presidents")
plt.show()
````
