## 7B: Random number simulation

```python
import numpy as np
import matplotlib.pyplot as plt


vx = []
vy = []
for i in range (2000):
    x = np.random.uniform(-10, 10)
    y = np.random.uniform(-10, 10)
    if (max(abs(x), abs(y)) > 5) and (np.sqrt(x**2 + y**2) < 10):
        vx.append(x)
        vy.append(y)
    

plt.plot(vx, vy, "b*")
plt.title("Sketch of the Cassiopeia star constalation")
plt.xlabel("relative x value from center star")
plt.ylabel("relative y value from center star")
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.grid()
plt.show()
```

```python
import numpy as np
x = np.ceil(6 * np. random.rand(10))
plt.hist(x, 6)
plt.title("10 Dice rolls")
plt.show()

x = np.ceil(6 * np. random.rand(100))
plt.hist(x, 6)
plt.title("100 Dice rolls")
plt.show()

x = np.ceil(6 * np. random.rand(1000))
plt.hist(x, 6)
plt.title("1.000 Dice rolls")
plt.show()

x = np.ceil(6 * np. random.rand(2000))
plt.hist(x, 6)
plt.title("2.000 Dice rolls")
plt.show()

x = np.ceil(6 * np. random.rand(4000))
plt.hist(x, 6)
plt.title("4.000 Dice rolls")
plt.show()

x = np.ceil(6 * np. random.rand(8000))
plt.hist(x, 6)
plt.title("8.000 Dice rolls")
plt.show()

x = np.ceil(6 * np. random.rand(10000))
plt.hist(x, 6)
plt.title("10.000 Dice rolls")
plt.show()

x = np.ceil(6 * np. random.rand(10000))
plt.hist(x, 6)
plt.title("50.000 Dice rolls")
plt.show()
```

At around 2.000 dice rolls the distribution starts looking uniform, but it never becomes perfectly uniform, even at 50.000 rolls.

We need more dice throws than coin flips, since there are 3 times as many possibly outcomes, and for a low N, there may be numbers that were never rolled with a dice.

## 7D: Carbon dating

```python
import numpy as np
def carbonTime(start, N_0):
    L = np.log(2) / 5700
    t = []
    fin = N_0 * 10
    for i in range (1, fin+1):
        N = i / 10
        time = - (1 / L) * np.log(N / N_0)
        t.append(time)
    return t
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = []
for i in range (1000):
    x.append(i / 10)
y = carbonTime(0, 100)
plt.plot(x, y, "r")
plt.title("Carbon-14 Decay")
plt.xlabel("Percentage of C-14 remaining")
plt.ylabel("Age of materal (years)")
plt.xlim([0, 100])
plt.ylim([0, 60000])
plt.grid()
plt.show()
```
