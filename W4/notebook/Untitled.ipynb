```python
import pandas as pd
dataframe = pd.read_csv("co2full.csv", sep = ",")
```

```python
avgs = dataframe['average']
dates = dataframe['date']
```

```python
avgs
```

```python
months = []
years = []
for x in dates:
    y = int(x)
    years.append(y)
    m = int((x - y) * 12) + 1
    months.append(m)
```

```python
for i in range(len(dates)):
    y,m,a = years[i], months[i], avgs[i]
    print('%i,%i,%f' % (y,m,a))
```
